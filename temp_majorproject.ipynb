```python
import torch
import torch.nn as nn

class AdaptiveMotionSaliencyScorer(nn.Module):

    def __init__(self, dim=512, keep_ratio=0.5):
        super().__init__()

        self.keep_ratio = keep_ratio

        self.score = nn.Sequential(
            nn.Linear(dim, dim // 2),
            nn.GELU(),
            nn.Linear(dim // 2, 1)
        )

    def forward(self, x):

        # x shape:
        # [B, T, N, C]

        B, T, N, C = x.shape

        scores = self.score(x).squeeze(-1)

        k = max(1, int(N * self.keep_ratio))

        topk_indices = scores.topk(k, dim=-1).indices

        topk_indices = topk_indices.unsqueeze(-1).expand(-1, -1, -1, C)

        pruned_x = torch.gather(x, 2, topk_indices)

        return pruned_x
```

```python
import torch
import torch.nn as nn

from models.modeling_motion_tokenizer import SpatioTemporalBlock


class AMSSSpatioTemporalBlock(SpatioTemporalBlock):

    def __init__(self, dim, num_heads, keep_ratio=0.5):
        super().__init__(
            dim=dim,
            num_heads=num_heads
        )

        self.amss = AdaptiveMotionSaliencyScorer(
            dim=dim,
            keep_ratio=keep_ratio
        )

    def forward(self, x):

        print("Before AMSS:", x.shape)

        x = self.amss(x)

        print("After AMSS:", x.shape)

        return super().forward(x)
```

```python
x = torch.randn(2, 24, 100, 512)

block = AMSSSpatioTemporalBlock(
    dim=512,
    num_heads=8,
    keep_ratio=0.5
)

out = block(x)

print("Final output shape:", out.shape)
```

```python
import time
import torch

# Fake tensor
x = torch.randn(2, 24, 100, 512).cuda()

# Baseline block
baseline_block = SpatioTemporalBlock(
    dim=512,
    num_heads=8
).cuda()

# AMSS block
amss_block = AMSSSpatioTemporalBlock(
    dim=512,
    num_heads=8,
    keep_ratio=0.5
).cuda()


# -----------------------------
# Baseline timing
# -----------------------------
torch.cuda.synchronize()

start = time.time()

for _ in range(20):
    out = baseline_block(x)

torch.cuda.synchronize()

baseline_time = time.time() - start


# -----------------------------
# AMSS timing
# -----------------------------
torch.cuda.synchronize()

start = time.time()

for _ in range(20):
    out = amss_block(x)

torch.cuda.synchronize()

amss_time = time.time() - start


print(f"Baseline Runtime: {baseline_time:.4f} sec")
print(f"AMSS Runtime: {amss_time:.4f} sec")

speedup = baseline_time / amss_time

print(f"Speedup: {speedup:.2f}x")
```

benchmark block

```python
class SilentAMSSSpatioTemporalBlock(SpatioTemporalBlock):

    def __init__(self, dim, num_heads, keep_ratio=0.5):
        super().__init__(
            dim=dim,
            num_heads=num_heads
        )

        self.amss = AdaptiveMotionSaliencyScorer(
            dim=dim,
            keep_ratio=keep_ratio
        )

    def forward(self, x):

        x = self.amss(x)

        return super().forward(x)
```

```python
import time
import torch

x = torch.randn(2, 24, 100, 512).cuda()

baseline_block = SpatioTemporalBlock(
    dim=512,
    num_heads=8
).cuda()

amss_block = SilentAMSSSpatioTemporalBlock(
    dim=512,
    num_heads=8,
    keep_ratio=0.5
).cuda()


# Warmup
for _ in range(5):
    baseline_block(x)
    amss_block(x)


# --------------------------
# Baseline
# --------------------------
torch.cuda.synchronize()

start = time.time()

for _ in range(50):
    baseline_block(x)

torch.cuda.synchronize()

baseline_time = time.time() - start


# --------------------------
# AMSS
# --------------------------
torch.cuda.synchronize()

start = time.time()

for _ in range(50):
    amss_block(x)

torch.cuda.synchronize()

amss_time = time.time() - start


speedup = baseline_time / amss_time

print(f"Baseline Runtime: {baseline_time:.4f} sec")
print(f"AMSS Runtime: {amss_time:.4f} sec")
print(f"Speedup: {speedup:.2f}x")
```

```python
import time
import torch

keep_ratios = [1.0, 0.75, 0.5, 0.25]

results = []

x = torch.randn(2, 24, 100, 512).cuda()

for ratio in keep_ratios:

    if ratio == 1.0:

        block = SpatioTemporalBlock(
            dim=512,
            num_heads=8
        ).cuda()

    else:

        block = SilentAMSSSpatioTemporalBlock(
            dim=512,
            num_heads=8,
            keep_ratio=ratio
        ).cuda()

    # Warmup
    for _ in range(5):
        block(x)

    torch.cuda.synchronize()

    start = time.time()

    for _ in range(50):
        block(x)

    torch.cuda.synchronize()

    runtime = time.time() - start

    output_tokens = int(100 * ratio)

    results.append((ratio, output_tokens, runtime))

    print(f"Keep Ratio: {ratio}")
    print(f"Output Tokens: {output_tokens}")
    print(f"Runtime: {runtime:.4f} sec")
    print("-" * 40)
```
